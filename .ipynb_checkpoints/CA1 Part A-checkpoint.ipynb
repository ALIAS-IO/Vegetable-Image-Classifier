## Data Analysis

```python
data_dir = pathlib.Path("./CA1-Part_A-Dataset/Dataset for CA1 part A - AY2526S1")
train_dir = data_dir / "train"
val_dir = data_dir / "validation"
test_dir = data_dir / "test"

batch_size = 32

ds_train = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    batch_size=batch_size
)

ds_test = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    batch_size=batch_size
)

ds_val = tf.keras.preprocessing.image_dataset_from_directory(
    val_dir,
    batch_size=batch_size
)
```

```python
def preprocess_and_resize(ds, size):
    return ds.map(lambda x, y: (tf.image.resize(tf.image.rgb_to_grayscale(x), size), y))
```

```python
ds_train_23 = preprocess_and_resize(ds_train, (23, 23))
ds_test_23 = preprocess_and_resize(ds_test, (23, 23))
ds_val_23 = preprocess_and_resize(ds_val, (23, 23))

ds_train_101 = preprocess_and_resize(ds_train, (101, 101))
ds_test_101 = preprocess_and_resize(ds_test, (101, 101))
ds_val_101 = preprocess_and_resize(ds_val, (101, 101))

class_names = ds_train.class_names

print(class_names)
```

Next, we will see what is the average image for each class label to see how similar each image is

```python
def plot_avg_images(dataset, class_names, n_rows=3, n_cols=5, cmap='gray', title=""):
    images = []
    labels = []
    for batch_images, batch_labels in dataset.unbatch():
        images.append(batch_images.numpy())
        labels.append(batch_labels.numpy())
    
    X = np.stack(images)
    y = np.array(labels)
    
    fig, ax = plt.subplots(n_rows, n_cols, figsize=(24, 10))
    plt.suptitle(title, fontfamily='serif', fontsize=20, fontweight='bold')

    for idx, subplot in enumerate(ax.ravel()):
        if idx >= len(class_names):
            subplot.axis('off')
            continue

        class_mask = y == idx
        class_images = X[class_mask]

        if len(class_images) > 0:
            avg_image = np.mean(class_images, axis=0)
            subplot.imshow(avg_image, cmap=cmap)
            subplot.set_title(class_names[idx], fontfamily='serif', fontsize=12, fontweight='bold')
            subplot.axis('off')
        else:
            subplot.axis('off')
```

```python
plot_avg_images(ds_train_23, class_names=class_names, title="23x23 Average Images")
plot_avg_images(ds_train_101, class_names=class_names, title="101x101 Average Images")
```

#### Observation:

- From the average image of each class label, it is not very clear to see and predict what vegetable it is
- The 101 by 101 seem to make it slightly clearer to see what some of the vegetables are but are still not very clear
- This suggests that there is quite a significant amount of variation within the images of all 15 classes of vegetables

## 23 x 23

### Baseline for 23x23 (Original)

```python
X_batches, y_batches = zip(*ds_train_23.as_numpy_iterator())
X_train = np.vstack(X_batches)
y_train = np.hstack(y_batches)

X_train_23 = X_train.astype('float32') / 255.0
y_train_23 = to_categorical(y_train)

print("X_train shape:", X_train_23.shape)
print("y_train shape:", y_train_23.shape)
```

```python
X_batches, y_batches = zip(*ds_val_23.as_numpy_iterator())

X_val = np.vstack(X_batches)
y_val = np.hstack(y_batches)

X_val_23 = X_val.astype('float32') / 255.0

y_val_23 = to_categorical(y_val)

print("X_val shape:", X_val_23.shape)
print("y_val shape:", y_val_23.shape)
```

```python
X_batches, y_batches = zip(*ds_test_23.as_numpy_iterator())

X_test = np.vstack(X_batches)
y_test = np.hstack(y_batches)

X_test_23 = X_test.astype('float32') / 255.0

y_test_23 = to_categorical(y_test)

print("X_test shape:", X_test_23.shape)
print("y_test shape:", y_test_23.shape)
```

```python
num_classes = y_test_23.shape[1]
```

```python
OriginalModel23 = Sequential([
    Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=(23, 23, 1), kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Conv2D(128, (3, 3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Flatten(),
    Dropout(0.5),

    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    Dense(num_classes, activation='softmax')
])

OriginalModel23.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint = ModelCheckpoint('./Part A checkpoints/original_model_weights_23.h5', monitor='val_accuracy', save_best_only=True, save_weights_only=True, mode='max', verbose=1)

history = OriginalModel23.fit(X_train_23, y_train_23, validation_data=(X_val_23, y_val_23), epochs=100, batch_size=32, callbacks=[checkpoint], verbose=1)

OriginalModel23.summary()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

fig.suptitle('Original Performance', fontsize=16)

axs[0].plot(history.history['accuracy'])
axs[0].plot(history.history['val_accuracy'])
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Train', 'Validation'], loc='upper left')

axs[1].plot(history.history['loss'])
axs[1].plot(history.history['val_loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

### Balanced Data Augmentation 

```python
X_batches, y_batches = zip(*ds_train_23.as_numpy_iterator())

X_train = np.vstack(X_batches)
y_train = np.hstack(y_batches)

X_train = X_train.astype('float32') / 255.0

class_counts = Counter(y_train)
max_count = max(class_counts.values())

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

X_balanced = []
y_balanced = []

for label in class_counts:
    idx = np.where(y_train == label)[0]
    X_class = X_train[idx]
    y_class = y_train[idx]

    X_balanced.extend(X_class)
    y_balanced.extend(y_class)

    num_to_generate = max_count - len(idx)
    if num_to_generate > 0:

        aug_iter = datagen.flow(X_class, y_class, batch_size=1)
        for _ in range(num_to_generate):
            x_aug, y_aug = next(aug_iter)
            X_balanced.append(x_aug[0])
            y_balanced.append(y_aug[0])

X_train_balanced_23 = np.array(X_balanced)
y_train_balanced_23 = to_categorical(np.array(y_balanced))

class_names = ds_train.class_names
original_counts = Counter(y_train)
augmented_counts = Counter(np.argmax(y_train_balanced_23, axis=1))

labels = sorted(original_counts.keys())
original_values = [original_counts[label] for label in labels]
augmented_values = [augmented_counts[label] for label in labels]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, original_values, width, label='Original')
plt.bar(x + width/2, augmented_values, width, label='Augmented')
plt.xlabel('Class Label')
plt.ylabel('Image Count')
plt.title('Image Count Per Class: Before and After Augmentation')
plt.xticks(x, class_names, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("Balanced X shape:", X_train_balanced_23.shape)
print("Balanced y shape:", y_train_balanced_23.shape)
```

```python
AugmentedModel23 = Sequential([
    Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=(23, 23, 1), kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Conv2D(128, (3, 3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Flatten(),
    Dropout(0.5),

    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    Dense(num_classes, activation='softmax')
])

AugmentedModel23.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint = ModelCheckpoint('./Part A checkpoints/augmented_model_weights_23.h5', monitor='val_accuracy', save_best_only=True, save_weights_only=True, mode='max', verbose=1)

history = AugmentedModel23.fit(X_train_balanced_23, y_train_balanced_23, validation_data=(X_val_23, y_val_23), epochs=100, batch_size=32, callbacks=[checkpoint], verbose=1)

AugmentedModel23.summary()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

fig.suptitle('Augemented Performance', fontsize=16)

axs[0].plot(history.history['accuracy'])
axs[0].plot(history.history['val_accuracy'])
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(['Train', 'Validation'], loc='upper left')

axs[1].plot(history.history['loss'])
axs[1].plot(history.history['val_loss'])
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

#### Observations:
- Less noisy, improvement over original
- Could still be improved

```python
model_name = "Augmented"

model_hist = {
    "Mean loss": np.mean(history.history['loss']),
    "Mean accuracy": np.mean(history.history['accuracy']),
    "Mean val_loss": np.mean(history.history['val_loss']),
    "Mean val_accuracy": np.mean(history.history['val_accuracy']),
}

model_df = pd.DataFrame([model_hist], index=[model_name])

perf_23_models = pd.concat([perf_23_models, model_df])

display(perf_23_models)
```

### Balanced Data Augmentation

```python
X_batches, y_batches = zip(*ds_train_101.as_numpy_iterator())

X_train = np.vstack(X_batches)
y_train = np.hstack(y_batches)

X_train = X_train.astype('float32') / 255.0

class_counts = Counter(y_train)
max_count = max(class_counts.values())

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.15,
    shear_range=5,
    horizontal_flip=True,
    fill_mode='nearest'
)



X_balanced = []
y_balanced = []

for label in class_counts:
    idx = np.where(y_train == label)[0]
    X_class = X_train[idx]
    y_class = y_train[idx]

    X_balanced.extend(X_class)
    y_balanced.extend(y_class)

    num_to_generate = max_count - len(idx)
    if num_to_generate > 0:
        aug_iter = datagen.flow(X_class, y_class, batch_size=1)
        for _ in range(num_to_generate):
            x_aug, y_aug = next(aug_iter)
            X_balanced.append(x_aug[0])
            y_balanced.append(y_aug[0])

X_train_balanced_101 = np.array(X_balanced)
y_train_balanced_101 = to_categorical(np.array(y_balanced))
```

```python
print("Balanced X shape:", X_train_balanced_101.shape)
print("Balanced y shape:", y_train_balanced_101.shape)
```

```python
AugmentedModel101 = Sequential([
    Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(101, 101, 1), kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Conv2D(64, (3, 3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Conv2D(128, (3, 3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    Conv2D(256, (3, 3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    MaxPooling2D(pool_size=2),

    GlobalAveragePooling2D(),
    Dropout(0.3),

    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    BatchNormalization(),
    Dropout(0.3),

    Dense(num_classes, activation='softmax')
])

AugmentedModel101.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint = ModelCheckpoint('./Part A checkpoints/augmented_model_weights_101.h5', monitor='val_accuracy', save_best_only=True, save_weights_only=True, mode='max', verbose=1)

history = AugmentedModel101.fit(X_train_balanced_101, y_train_balanced_101, validation_data=(X_val_101, y_val_101), epochs=100, batch_size=64, callbacks=[checkpoint], verbose=1)

AugmentedModel101.summary()
```
